```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name
```

```python
data = {
    'ชื่อ': ['นักศึกษา1', 'นักศึกษา2', 'นักศึกษา3', 'นักศึกษา4', 'นักศึกษา5',
             'นักศึกษา6', 'นักศึกษา7', 'นักศึกษา8', 'นักศึกษา9', 'นักศึกษา10'],
    'อายุ': [20, 21, 22, 19, 20, 150, 21, 22, 23, 18]  # 150 คือค่าผิดปกติ
}
```

กำหนดให้ทุกช่องว่างในคอลัมน์อายุ ถูกแทนที่ด้วยค่าเฉลี่ยอายุ (แก้ไขใน df ทันที)

```python
df = pd.DataFrame(data)
```

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

หา Q1 (25%) และ Q3 (75%) ของข้อมูลอายุใน df เพื่อใช้วิเคราะห์หรือกรองค่า outlier

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

```python
Q1, Q3
```

IQR คือค่าช่วงควอไทล์ คำนวณจาก Q3 ลบ Q1 เพื่อตรวจสอบช่วงกระจายตัวของข้อมูลส่วนกลาง


```python
IQR = Q3 - Q1
```

```python
IQR
```

ขอบเขตล่างและขอบเขตบนสำหรับระบุ outlier โดยใช้ควอไทล์และ IQR

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

ขอบเขตล่างและขอบเขตบนสำหรับระบุ outlier โดยใช้ควอไทล์และ IQR


```python
print(lower_bound)
print(upper_bound)
```

ตัดเอาเฉพาะแถวที่ข้อมูลอายุไม่ใช่ outlier (ไม่หลุดช่วงปกติ) มาเก็บใน df_no_outliers

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df_no_outliers
```

ตัดเอาเฉพาะแถวที่ข้อมูลอายุไม่ใช่ outlier (ไม่หลุดช่วงปกติ) มาเก็บใน df_no_outliers

```python
df['อายุ'] >= lower_bound
```

เช็คว่าค่า 'อายุ' ใน df น้อยกว่าหรือเท่ากับ upper_bound หรือไม่ (ผลลัพธ์เป็น True/False)

```python
df['อายุ'] <= upper_bound
```

สร้างตารางข้อมูลบันทึกยอดขายกาแฟในแต่ละวันและเวลาที่เฉพาะเจาะจง โดยใช้ datetime เพื่อการจัดการวันเวลาได้อย่างมีประสิทธิภาพ

```python
coffee_data = pd.DataFrame({
    'วันเวลา': pd.to_datetime(['2024-01-15 07:30', '2024-01-15 12:00',
                               '2024-01-15 16:30', '2024-01-16 08:00']),
    'ยอดขาย': [1500, 3200, 2100, 1800]
})
```

```python
coffee_data
```

เพิ่มคอลัมน์ 'ชั่วโมง' ที่บอกว่าข้อมูลแต่ละแถวถูกบันทึกไว้ในชั่วโมงที่เท่าไรของวัน (0–23)

```python
coffee_data['ชั่วโมง'] = coffee_data['วันเวลา'].dt.hour
```

```python
coffee_data
```

เพิ่มคอลัมน์ชื่อวันของสัปดาห์ให้ DataFrame โดยดึงจากข้อมูลวันเวลาแต่ละแถว

```python
coffee_data['วันในสัปดาห์'] = coffee_data['วันเวลา'].dt.day_name()
```

```python
coffee_data
```

แปลงชั่วโมงของแต่ละข้อมูลให้อยู่ในช่วงของวัน (เช้า/กลางวัน/บ่าย/เย็น) เพื่อสะดวกต่อการวิเคราะห์ข้อมูลยอดขายในแต่ละช่วงเวลา

```python
coffee_data['ช่วงเวลา'] = pd.cut(coffee_data['ชั่วโมง'],
                                  bins=[0, 10, 14, 18, 24],
                                  labels=['เช้า', 'กลางวัน', 'บ่าย', 'เย็น'])
```

```python
coffee_data
```

```python
print(coffee_data[['วันเวลา', 'ช่วงเวลา', 'ยอดขาย']])
```

เป็นการสร้างข้อมูลพื้นฐานนักเรียนเพื่อใช้วิเคราะห์เกี่ยวกับสุขภาพ เช่น สามารถนำไปคำนวณค่า BMI ได้ต่อไป

```python
student_health = pd.DataFrame({
    'ชื่อ': ['สมชาย', 'สมหญิง', 'สมศักดิ์'],
    'ส่วนสูง_cm': [170, 165, 175],
    'น้ำหนัก_kg': [65, 55, 80]
})
```

เพิ่มคอลัมน์ที่คำนวณค่าดัชนีมวลกาย (BMI) และแปะป้ายสุขภาพ (ผอม/ปกติ/อ้วน/อ้วนมาก) ให้แต่ละคนใน DataFrame student_health

```python
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```

```python
print(student_health[['ชื่อ', 'BMI', 'สุขภาพ']])
```

faculty_mapping เป็น dictionary ที่ใช้สำหรับแมป (mapping) หรือเปลี่ยนชื่อคณะใน df_student ให้มีรูปแบบเดียวกันเพื่อความถูกต้องและสะดวกในการวิเคราะห์ข้อมูล


```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์'
}
```

```python
faculty_mapping
```

คำสั่งนี้จะใช้ dictionary faculty_mapping เพื่อเปลี่ยนชื่อคณะในคอลัมน์ df_student['คณะ'] ให้เป็นไปตามมาตรฐานที่กำหนดไว้
โดยใช้ .map() กับคอลัมน์ "คณะ" ใน DataFrame

```python
df_student['คณะ'] = df_student['คณะ'].map(faculty_mapping)
```
